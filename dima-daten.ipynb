Playing with similarity: Given an entry, eventually one likes to get a list of the most similar entries.

```python
import spacy
# nlp = spacy.load('en_core_web_sm')
nlp = spacy.load('de_core_news_sm')

# def similar_entries(id, count):
def similar_entries(id):
    conn = get_db_connection()
    reference_entry = conn.execute('SELECT * FROM entries WHERE id = ?', (id,)).fetchone()
    print('---')
    if not reference_entry:
        print('id has no entry')
        return 
    
    conn.close()
    sim_list = []
    id_list = []
    
    for entry in entries:
        sim = difflib.SequenceMatcher(None, entry['content'], reference_entry['content']).ratio()

        doc1 = nlp( entry['content'] )
        doc2 = nlp( reference_entry['content'] )
        sim = doc1.similarity( doc2 )
        
        sim_list.append(sim)
        id_list.append(entry['id'])
        sim_id_list = dict(zip(sim_list, id_list)) # convert lists to dictionary
        # print('.')
    return sim_id_list
        
sim_id_list = similar_entries('199')
```

```python
# generate id_list of entries where keyword appears
def get_id_list(searchword):
    id_list = []
    entries = get_all_entries() # reload
    for entry in entries:
        if searchword in entry['content']:
            id_list.append(entry['id'])
            # print(entry['id'],': ', entry['content'], '\n')
    return id_list


def db_add_searchwords(id_list, searchword):
    conn = get_db_connection()

    for id in id_list:
        # conn = get_db_connection()    
        entry = conn.execute('SELECT * FROM entries WHERE id = ?', (id,)).fetchone()
        # conn.close()
        searchwords = entry['searchwords']

        if searchwords == None:
            searchwords = searchword
        else:
            searchwords = searchwords + ', ' + searchword
        # print(keywords)
        # conn = get_db_connection()
        conn.execute('UPDATE entries SET searchwords = ? WHERE id = ?', (searchwords, id))
        conn.commit()
        # conn.close()
    conn.close()
    
    
searchword = 'teuer'
id_list = get_id_list(searchword)
    
print(id_list)
        
```
